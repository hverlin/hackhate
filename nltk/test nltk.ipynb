```python
import nltk
import pandas as pd
from nltk.corpus import stopwords
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()


stopwords = set(stopwords.words('english'))
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'[a-z][a-z]+')
```

```python
with open('prefixes.txt', 'r') as f:
    prefixes = f.read()
prefixes = prefixes.split("\n")[:-1]
```

```python
raw_data = pd.read_csv("training.1600000.processed.noemoticon.csv")
```

```python
data = raw_data.drop(raw_data.columns[[1, 2, 3, 3, 4]], axis=1)
```

```python
data.head()
```

```python
data['score'] = 0
```

```python
for line in data.tail(100).iterrows():
    txt  = list(line[1])[1]
    print(txt, '\n', sid.polarity_scores(text=txt), "\n")
```

```python
data_neg = data.head(1000)
data_pos = data.tail(1000)
```

```python
def filter_words(word:str, prefixes):
    for w in prefixes:
        if w in word:
            return False
    return True
```

```python
neg_matrix = data_neg.as_matrix()
for d in neg_matrix:
    d[1] = " ".join([w for w in d[1].split() if filter_words(w, prefixes=prefixes)])
    d[1] = tokenizer.tokenize(d[1])
    d[1] = [w.lower() for w in d[1] if w.lower() not in stopwords]
```
